```python
#take in a pair of coordinate pairs in lat, lon and return distance in km
def distance(point1, point2):
    return geopy.distance.geodesic(point1, point2).km

#convert longitudes in (0 - 360) to (-180 - 180)
def lon_convert(x):
    if x > 180:
        return x - 360
    else:
        return x

#WPR relation
def pressure_to_wind(pressure):
    if pressure >= 1010:
        return 0
    
    return 2.3*((1010-pressure)**(.76))

#Current trajectory stitching prioritizes age in assigning storm instances to existing storms,
#but you could try intensity-based priority as well by including this function
def sort_storms(living_storms):
    #insertion sort to make sure biggest living storms maintain continuity over smaller ones
    #sorted by pressure min
    for i in range(1, len(living_storms)):
        for j in range(i, 0, -1):
            if living_storms[j]["pressure_min"] < living_storms[j-1]["pressure_min"]:
                hold = living_storms[j]
                living_storms[j] = living_storms[j - 1]
                living_storms[j - 1] = hold
            else:
                break

#for making charts
def assign_cat(pressure):
        max_wind = pressure_to_wind(pressure)
        if max_wind > 70:
            return 5
        elif max_wind > 58:
            return 4
        elif max_wind > 50:
            return 3
        elif max_wind > 43:
            return 2
        elif max_wind > 33:
            return 1
        else:
            return 0
```

```python
###DETECT STORMS
#Change file names as needed

for year in ["2090", "2091", "2092", "2093", "2094", "2095", "2096", "2097", "2098", "2099"]:

    name = "data/cam.h1/" + year + "/combined.nc"
    file = ncds(name)

    name = "data/cam.t/" + year + "/combined.nc"
    file2 = ncds(name)

    lon = np.array(file.variables["lon"])
    lat = np.array(file.variables["lat"])
    vorticity = np.absolute(file.variables["vorticity850"])
    ps = (file.variables["PSL"]) #CHANGES BETWEEN 80 AND 94
    u10 = file.variables["U10"]
    t = file.variables["T300-500"] #CHANGES BETWEEN 80 AND 94

    TS = np.array(file2.variables["TS"])

    #minimum vorticity for a TC, given by Zhao et al.
    vort_threshold = 1.6 * 10**-4
    time_length = len(file.variables["time"])

    height = len(lat)
    width = len(lon)
    window_6 = 24 #cell radius to go out 6 degrees
    window_2 = 8 #cell radius to go out 2 degrees
    window_4 = 16 #cell radius to go out 4 degrees
    window_1 = 4 #cell radius to go out 1 degree
    window_1_5 = 6 #1.5 degrees
    window_3 = 12 #3 degrees

    #array to story dicts of instantaneous storm detections and their properties
    storm_data = [None] * time_length
    TC_data = [None] * time_length

    #iterate through every 6 hour step
    for time in range(time_length):

        candidate_count = 0
        print(time, "/", time_length)

        #make arrays for the sim values at each time step
        PSL = (ps[time])
        averaged_temp = (t[time])
        vorticities = (np.array(vorticity[time]))
        u = (u10[time])
        surface_TS = TS[time]

        #specific to our model runs
        for i in range(vorticities.shape[0]):
            for j in range(vorticities.shape[1]):
                if vorticities[i][j] > 100: #remove default values
                    vorticities[i][j] = 0

        #find TC's at each time step
        storm_info = []
        TC_info = []
        for i in range(vorticities.shape[0]):
            for j in range(vorticities.shape[1]):
                not_lowest = False

                #make sure a vorticity is above the threshold, and its the highest among itself and those in its
                #proximity 2 degrees area
                #CHANGED FROM VITART
                if vorticities[i][j] > vort_threshold:
                    for k in range(max(0, i-window_2), min(height, i+window_2)):
                        for l in range((j-window_2), (j+window_2)):
                            if not_lowest:
                                break
                            if l >= width or l < 0:
                                l = l%width #wrap around globe
                            if (((abs(lon[j] - lon[l]) <= 2) or (abs(lon[j] - lon[l]) >= 358)) and abs(lat[i] - lat[k]) <= 2):
                                if (vorticities[k][l] > (vorticities[i][j])):
                                    not_lowest = True

                    #happens if we are indeed above the threshold vorticity, and at a local vorticity maxiumum
                    if not_lowest == False:
                        candidate_count += 1
                        pressure_min = float('inf')
                        lowest_PSL_index = None

                        #scan through all the PSL's of the cells in the proximities 2 degree area
                        for k in range(max(0, i-window_2), min(height, i+window_2)):
                            for l in range((j-window_2), (j+window_2)):
                                if l >= width or l < 0:
                                    l = l%width #wrap around globe
                                if (((abs(lon[j] - lon[l]) <= 2) or (abs(lon[j] - lon[l]) >= 358)) and abs(lat[i] - lat[k]) <= 2):
                                    if PSL[k][l] < pressure_min:
                                        pressure_min = PSL[k][l]
                                        lowest_PSL_index = (k, l)

                        #find maximum local wind within 2 degrees
                        wind_max = 0.0
                        highest_wind_index = None

                        #scan through all the wind speeds of the cells in the proximities 2 degree area around PSL min
                        for k in range(max(0, lowest_PSL_index[0]-window_2), min(height, lowest_PSL_index[0]+window_2)):
                            for l in range((lowest_PSL_index[1]-window_2), (lowest_PSL_index[1]+window_2)):
                                if l >= width or l < 0:
                                    l = l%width #wrap around globe
                                if (((abs(lon[lowest_PSL_index[1]] - lon[l]) <= 2) or (abs(lon[lowest_PSL_index[1]] - lon[l]) >= 358)) and abs(lat[lowest_PSL_index[0]] - lat[k]) <= 2):
                                    if u[k][l] > wind_max:
                                        wind_max = u[k][l]
                                        highest_wind_index = (k, l)

                        #find warm core center, the warmest cell within 4 degrees; TC is only valid if this center
                        #is within 2 degrees of storm center, and at least 1C warmer than 4 degree square around such
                        #CHANGED FROM VITART
                        temp_max = 0
                        highest_temp_index = None

                        #CHANGED LAST NIGHT
                        for k in range(max(0, lowest_PSL_index[0]-window_4), min(height, lowest_PSL_index[0]+window_4)):
                            for l in range((lowest_PSL_index[1]-window_4), (lowest_PSL_index[1]+window_4)):
                                if l >= width or l < 0:
                                    l = l%width #wrap around globe
                                if (((abs(lon[lowest_PSL_index[1]] - lon[l]) <= 4) or (abs(lon[lowest_PSL_index[1]] - lon[l]) >= 356)) and abs(lat[lowest_PSL_index[0]] - lat[k]) <= 4):
                                    if averaged_temp[k][l] > temp_max:
                                        temp_max = averaged_temp[k][l]
                                        highest_temp_index = (k, l)

                        #make sure our maximum height averaged temp within 2 degrees is at least a degree warmer than the 4 degree box mean around itself
                        #exception when there's another low pressure spot close by
                        #CHANGED FROM VITART
                        #print((max(0, highest_temp_index[0]-window_4), min(height, highest_temp_index[0]+window_4)), range((highest_temp_index[1]-window_4), (highest_temp_index[1]+window_4)%width))
                        if (((abs(lon[lowest_PSL_index[1]] - lon[highest_temp_index[1]]) <= 2) or (abs(lon[lowest_PSL_index[1]] - lon[highest_temp_index[1]]) >= 358)) and abs(lat[lowest_PSL_index[0]] - lat[highest_temp_index[0]]) <= 2) \
                                    and (temp_max >= np.mean([averaged_temp[m][n] \
                                    for m in range(max(0, highest_temp_index[0]-window_4), min(height, highest_temp_index[0]+window_4)) \
                                    for n in range((highest_temp_index[1]-window_4), (highest_temp_index[1]+window_4)%width)]) \
                                    + 1):
                            #only if there's a valid warm core
                            storm_info.append({"storm_center": (lowest_PSL_index), "lat": [lat[lowest_PSL_index[0]]], "lon": [lon[lowest_PSL_index[1]]], "time": [file.variables["time"][time]], "wind_max": [wind_max], "age": 1, "pressure_min": [pressure_min], "center_TS": [surface_TS[lowest_PSL_index[0]][lowest_PSL_index[1]]]})
                            #print({"storm_center": (lowest_PSL_index), "lat": lat[lowest_PSL_index[0]], "lon": lon[lowest_PSL_index[1]], "time": [file.variables["time"][time]], "wind_max": [wind_max], "age": 1})
                        #print((max(0, highest_temp_index[0]-window_4), min(height, highest_temp_index[0]+window_4)), range((highest_temp_index[1]-window_4), (highest_temp_index[1]+window_4)%width))

        print(len(storm_info), candidate_count)
        storm_data[time] = (storm_info)
        TC_data[time] = (TC_info)

    np.save("data/" + year + "_intervention_storm_data.npy", np.array(storm_data, dtype=object))
```

```python
storm_data = np.concatenate([np.load('data/2090_intervention_storm_data.npy', allow_pickle=True), np.load('data/2091_intervention_storm_data.npy', allow_pickle=True), np.load('data/2092_intervention_storm_data.npy', allow_pickle=True), np.load('data/2093_intervention_storm_data.npy', allow_pickle=True), np.load('data/2094_intervention_storm_data.npy', allow_pickle=True), np.load('data/2095_intervention_storm_data.npy', allow_pickle=True), np.load('data/2096_intervention_storm_data.npy', allow_pickle=True), np.load('data/2097_intervention_storm_data.npy', allow_pickle=True), np.load('data/2098_intervention_storm_data.npy', allow_pickle=True), np.load('data/2099_intervention_storm_data.npy', allow_pickle=True)])
snapshots = np.copy(storm_data)
```

```python
###MAKE TRAJECTORIES

#our algo here depends on having a list of storms that are born and die based on "snapshots," running detections
dates = np.arange(0, len(storm_data) * 6, 6)
living_storms = []
dead_storms = []
TC_count_by_day = []
TC_count_by_month = []

TS_threshold = 10
age_threshold = 12

#iterate through the snapshots, the instantaneous detections at each time step, and at each step make them all
#into continuations of existing storms, or new storms
for time in range(0, len(snapshots)):
    print(time, "/", len(snapshots))
    
    i = 0
    while i < (len(living_storms)):
        closest_distance = float("inf")
        closest_index = None
        #for each living storm, find the closest detected storm in the current snapshot
        for j in range(len(snapshots[time])):
            if distance([living_storms[i]["lat"][-1], living_storms[i]["lon"][-1]], [snapshots[time][j]["lat"][-1], snapshots[time][j]["lon"][-1]]) < closest_distance:
                closest_distance = distance((living_storms[i]["lat"][-1], living_storms[i]["lon"][-1]), (snapshots[time][j]["lat"][-1], snapshots[time][j]["lon"][-1]))
                closest_index = j
            #if two detections are equidistant, tie-break as per Zhao algo
            elif distance([living_storms[i]["lat"][-1], living_storms[i]["lon"][-1]], [snapshots[time][j]["lat"][-1], snapshots[time][j]["lon"][-1]]) == closest_distance:
                #if storms in snapshot are equidistant, rank by sum of diff in westwardness and polewardness
                if snapshots[time][j]["lon"][-1] - abs(snapshots[time][j]["lat"][-1]) < snapshots[time][closest_index]["lon"][-1] - abs(snapshots[time][closest_index]["lat"][-1]):
                    closest_distance = distance((living_storms[i]["lat"][-1], living_storms[i]["lon"][-1]), (snapshots[time][j]["lat"][-1], snapshots[time][j]["lon"][-1]))
                    closest_index = j
        #if a valid continuation has been found, take the living storm one step forward and update properties
        if len(snapshots[time]) > 0 and distance([living_storms[i]["lat"][-1], living_storms[i]["lon"][-1]], [snapshots[time][closest_index]["lat"][-1], snapshots[time][closest_index]["lon"][-1]]) < 400:
            lifetime = living_storms[i]["age"]
            wind_history = living_storms[i]["wind_max"]
            pressure_history = living_storms[i]["pressure_min"]
            
            TS_history = living_storms[i]["center_TS"]
            
            times = living_storms[i]["time"]
            
            lats = living_storms[i]["lat"]
            lons = living_storms[i]["lon"]
            
            living_storms[i] = snapshots[time].pop(closest_index) #stops from double counting
            
            living_storms[i]["age"] = lifetime + 1
            living_storms[i]["wind_max"] = wind_history + living_storms[i]["wind_max"]
            living_storms[i]["time"] = times + living_storms[i]["time"]
            living_storms[i]["pressure_min"] = pressure_history + living_storms[i]["pressure_min"]
            living_storms[i]["center_TS"] = TS_history + living_storms[i]["center_TS"]
            
            living_storms[i]["lat"] = lats + living_storms[i]["lat"]
            living_storms[i]["lon"] = lons + living_storms[i]["lon"]
        #if no continuation is found, storm dies, and we tally up whether it was old enough and consistently
        #strong enough to count as a valid TC
        else:
            #dead_storms is just the living storm dict, a bool for whether it reached the valid TC benchmark, and date of death
            dead_storms.append([living_storms.pop(i), False, dates[time]])
            if dead_storms[-1][0]["age"] >= age_threshold:
                days_over_17 = 0
                already_today = False
                for step in range(dead_storms[-1][0]["age"]):
                    if dead_storms[-1][0]["time"][step] % 24 == 0 or dead_storms[-1][0]["time"][step] == 0:
                        already_today = False
                    
                    if dead_storms[-1][0]["wind_max"][step] >= TS_threshold and pressure_to_wind((dead_storms[-1][0]["pressure_min"][step])/100) >= 17 and already_today == False:
                        days_over_17 += 1
                        already_today = True
                #make sure storm meets TC temp and wind persistence reqs
                if days_over_17 >= 3 and max(dead_storms[-1][0]["center_TS"]) >= 25.5 + 273.15:
                    dead_storms[-1][1] = True
            i -= 1 #this makes sure that, in popping from living storms, we don't skip over anything
        i += 1
    
    #add all remaining storms in snapshots to living storm list
    j = 0
    while j < (len(snapshots[time])):
        if abs(snapshots[time][j]["lat"][0]) < 50:
            living_storms.append(snapshots[time].pop(j)) #valid TC has to start below 50 degrees north or south
        else:
            snapshots[time].pop(j)

np.save('data/2090s_intervention_dead_storms.npy', np.array(dead_storms, dtype=object))
```
